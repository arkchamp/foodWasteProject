```python
#Connection
#using with to close the connection automatically
with psycopg.connect(
    dbname = "foodwaste",
    user = "postgres",
    password = "admin",
    host="localhost",
    port = "5432"
) as connection:
    with connection.cursor() as cur:

        #Function to insert data
        def insert_data(df, table_name):
            #from dataframe getting column names 
            columns = ",".join(df.columns)      #it is a pandas index -> pandas.index; it is iterable and behaves like a list of strings.
            placeholders = ",".join(["%s"] * len(df.columns))   # multiplying %s with the no. of columns
            query = f"insert into {table_name} ({columns}) values ({placeholders})"

            
            for row in df.itertuples(index=False, name=None):   #returns a row as a tuple
                try:
                    cur.execute(query,row)
                except Exception as e:
                    print(f"Error inserting into {table_name}:{e}")
                    connection.rollback() #If error occurs then Reset the transaction so next queries can run.

        # now inserting all the data
        insert_data(providers, "providers")
        insert_data(receivers, "receivers")
        insert_data(food, "food_listings")
        insert_data(claims, "claims")

        # commit the changes
        connection.commit()

print("Data inserted successfully.")
```
